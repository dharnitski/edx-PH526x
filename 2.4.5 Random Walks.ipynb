```python
import numpy as np
import matplotlib.pyplot as plt

# dot walks
delta_X = np.random.normal(0, 1, (2,10000))
# locations of dot
X = np.cumsum(delta_X, axis = 1)
# initial location
X_0 = np.array([[0], [0]])
# add initial dot to locations
X = np.concatenate((X_0, X), axis=1)
plt.plot(X[0], X[1], "ro-");
```
